```python
import sys
import os
import yaml

sys.path.append("/Users/daveih/Documents/python/usdm_data")

program_path = os.path.abspath("visualize.ipynb")

def read_yaml_file(file_path):
  with open(file_path, "r") as stream:
    return yaml.safe_load(stream)

def read_nodes(study, view=''):
  return read_yaml_file(full_file_path(study, 'nodes', view))

def read_edges(study, view=''):
  return read_yaml_file(full_file_path(study, 'edges', view))

def full_file_path(study, type, view):
  suffix = ''
  if view == 'timeline':
    suffix = '_timeline'
  return os.path.join(os.path.dirname(program_path), f"source_data/{study}{suffix}_{type}.yaml")
  
studies = [
  'Roche_NCT04320615_COVID', # Index 0
  'cycles_1',
  'cycles_2',
  'simple_1',
  'simple_2',
  'profile_1', # Index 5
  'simple_3',
  'CDISC_Pilot_Study',
  'EliLilly_NCT03421379_Diabetes'
]

study = studies[5]
```

```python
from yfiles_jupyter_graphs import GraphWidget

config = read_yaml_file(file_path = os.path.join(os.path.dirname(program_path), "config_data/yfiles_config.yaml"))
colour_map = config['colour_map']
shape_map = config['shape_map']

def custom_node_color(index: int, node: dict):
  if '_type' in node['properties'] and node['properties']['_type'] in colour_map:
    result = colour_map[node['properties']['_type']]
    return result
  else: 
    return 'white'

def custom_node_style(index: int, node: dict):
  if '_type' in node['properties'] and node['properties']['_type'] in shape_map:
    return { 'shape': shape_map[node['properties']['_type']] }
  else: 
    return { 'shape': 'ellipse' }
```

```python
nodes = read_nodes(study)
edges = read_edges(study)

graph_widget_1 = GraphWidget()
graph_widget_1.hierarchic_layout()
graph_widget_1.set_directed(True)

graph_widget_1.set_nodes(nodes)
graph_widget_1.set_edges(edges)
graph_widget_1.set_node_color_mapping(custom_node_color)
graph_widget_1.set_node_styles_mapping(custom_node_style)
graph_widget_1
```
